```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

##Grupo 1

```python
grupo1 = pd.read_csv("grupo1.csv")
grupo1.head()
```

```python
# Histograma
plt.figure(figsize=(6,4))
sns.histplot(grupo1["Calificacion"], bins=10, kde=True, color='skyblue')
plt.title("Grupo 1 - Histograma de Calificaciones")
plt.xlabel("Calificacion")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

```python
# Regresin lineal
X1 = grupo1[["HorasEstudio"]].values
y1 = grupo1["Calificacion"].values
modelo1 = LinearRegression().fit(X1, y1)
pred1 = modelo1.predict(X1)

# Parmetros
pendiente1 = modelo1.coef_[0]
intercepto1 = modelo1.intercept_
r1, _ = pearsonr(grupo1["HorasEstudio"], grupo1["Calificacion"])
r2_1 = r2_score(y1, pred1)

print(f"Ecuacin: y = {pendiente1:.2f}x + {intercepto1:.2f}")
print(f"Coeficiente de correlacin de Pearson (r): {r1:.4f}")
print(f"Coeficiente de determinacin (R): {r2_1:.4f}")
```

```python
# Dispersin + recta
plt.figure(figsize=(6,4))
sns.scatterplot(x=grupo1["HorasEstudio"], y=grupo1["Calificacion"])
plt.plot(grupo1["HorasEstudio"], pred1, color='red', label='Lnea de regresin')
plt.title("Grupo 1 - Regresin Lineal")
plt.xlabel("HorasEstudio")
plt.ylabel("Calificacion")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Coeficiente de correlacin
print(f"Coeficiente de correlacin (Pearson) para Grupo 1: r = {r1:.4f}")
```

**Explicar cul es la interpretacin del coeficiente de regresin en trminos de la
relacin entre el tiempo de estudio y las calificaciones finales?**

En este grupo, por cada hora adicional que un estudiante dedica al estudio, su calificacin final aumenta aproximadamente 2.47 puntos. Es decir, hay un efecto muy claro y marcado del estudio sobre la mejora en el rendimiento acadmico.

**Cmo se interpreta el coeficiente de correlacin en trminos de la fuerza y
direccin de la relacin entre las variables?**

El valor de la correlacin es 0.95, lo que indica una relacin muy fuerte y positiva. Esto quiere decir que los estudiantes que estudiaron ms, en general, obtuvieron mejores calificaciones, y la relacin entre ambas variables es muy consistente.

**El modelo de regresin lineal es estadsticamente significativo?**

S, el modelo es muy significativo. El valor de R fue 0.91, lo que significa que el modelo explica el 91% de las variaciones en las calificaciones. Esto indica que hay una relacin clara y confiable entre el tiempo de estudio y las calificaciones.

**Qu tan confiable es el modelo de regresin para predecir las
calificaciones finales de los estudiantes basado en el tiempo de estudio?**

Es muy confiable. Con un R tan alto y una correlacin fuerte, el modelo permite hacer buenas predicciones sobre la calificacin que podra obtener un estudiante en funcin de cunto estudia.
