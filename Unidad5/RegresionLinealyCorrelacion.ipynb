```python
from sklearn.linear_model import LinearRegression
import numpy as np

periodo = [2018, 2019, 2020, 2021, 2022, 2023]
inversion = np.array([2, 3, 4, 5, 11, 5]).reshape(-1, 1)
ganancias= [20, 25, 34, 30, 40, 31]

modelo = LinearRegression()
modelo.fit(inversion, ganancias)
ganancia_2024 = modelo.predict([[13]])
print("La ganancia esperada para el ao 2024 es de: ", ganancia_2024[0])
print("Coeficiente de la pendiente (m):", modelo.coef_[0])
print("Interpretacin de la recta (b): ", modelo.intercept_)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
inversion = [2, 3, 5, 4, 11, 5]
ganancias = [20, 25, 34, 30, 40, 31]
sns.set(style="whitegrid")
sns.regplot(x=inversion, y=ganancias, scatter=True, color='blue')
plt.title('Diagrama de Dispersin con linea de regresin')
plt.xlabel('Inversion (millones)')
plt.ylabel('Ganancias (millones)')
plt.grid(True)
plt.show()
```

```python
inversion = np.array([2, 3, 5, 4, 11, 5]).reshape(-1, 1)
inversion_prediccion = np.array([[8]])
ganancia_predicha = modelo.predict(inversion_prediccion)

plt.figure(figsize=(6, 5))
plt.scatter(inversion, ganancias, color='blue', label='Datos')
plt.plot(inversion, modelo.predict(inversion), color='red', label='Linea de regresin')
plt.scatter(inversion_prediccion, ganancia_predicha, color='green', marker='x', s=100,
        label='Prediccin (8, (0:,2f))'.format (ganancia_predicha[0]))
plt.legend()
```

```python
from scipy.stats import pearsonr

# Datos de entrada
publicidad = [2, 3, 5, 4, 11, 5]
ganancias = [20, 25, 34, 30, 40, 31]

# Calcular el Coeficiente de correlacin de Pearson
coeficiente_correlacion, p_valor = pearsonr(publicidad, ganancias)

print("Coeficiente de correlacin de Pearson;", coeficiente_correlacion)
print("Valor p:", p_valor)
```
