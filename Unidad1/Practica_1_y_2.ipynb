```python
from statistics import *
```

```python
# Lista de calififcaciones del grupo a de probabilidad y estadistica

grupoa = [70, 70, 65, 90, 100, 95, 45, 80, 73, 20, 100, 96]
print(grupoa)
```

#Determinar la mediana del ejemplo 1

```python
#Determinar el valo de la mediana con la función median
mediana = median(grupoa)
print(mediana)
```

#Determinar la moda del ejemplo 1

```python
#Determinar el valor de la moda con la función mode

moda = mode(grupoa)
print(moda)
```

```python
# En cado de que exista más de una moda, se puede emplear la función multimode para sacar las multiples modas que puedan existir

moda2 = multimode(grupoa)
print(moda2)
```

#Ejemplo 2. Tiempo de 100 metros de un grupo de 12 alumnos.

```python
#Definimos una variable llamada tiempos10mts de tipo lista, que contiene los tiempos que tardaron los 10 alumnos para correr los 100 metros
tiempo100mts = [15.1, 17.2, 14.69, 13.27, 22.15, 18.71, 19.15, 20.65, 15.10, 17.20]

#Desplegar los valores de la lista
print(tiempo100mts)
```

#Función mean para calcular el promedio

```python
#Usamos la función mean para determinar el tiempo promedio de la lista de valores

print("El tiempo promedio de los corredores es =>", mean(tiempo100mts))
```

#Función median para calcular la mediana

```python
#Usamos la función median para determinar el tiempo promedio de la lista de valores

#Definimos una variable llamada mediana para calcular la mediana de la lista con la función median
mediana = median(tiempo100mts)
print("La media de los tiempos de los corredores es =>", mediana)
```

#Funcion mode para calcular la moda

```python
#Definimos una variable llamda moda para calcular la moda de la lista con la función mode

moda = mode(tiempo100mts)
print("La moda de los tiempos es =>" , moda)
```

```python
multimodas = multimode(tiempo100mts)
print("La moda de los tiempos es => ", multimodas)
```

#Ejemplos 3: Histograma de 2 grupos

```python
#Libreria para generar números aleatorios
import random
```

```python
fabrica1 = []
for i in range (100):
  nAl = round(random.uniform(50, 100), 2)
  fabrica1.append(nAl)
print(fabrica1)
```

```python
fabrica2 = []
for i in range (100):
  nAl = round(random.uniform(50, 100), 2)
  fabrica2.append(nAl)
print(fabrica2)
```

```python
# libreria de python para graficar valores
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Producción diaria de la Fabrica 1")
```

```python
# libreria de python para graficar valores
import matplotlib.pyplot as plt2
```

```python
plt2.hist(fabrica2)
plt2.title("Producción diaria de la Fabrica 2")
```

```python
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)

print("Valores de la fabrica 1:")
print("La media es = ", media1)
print("La mediana es = ", mediana1)
print("La moda es = ", moda1)
```

```python
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)

print("Valores de la fabrica 2:")
print("La media es = ", media2)
print("La mediana es = ", mediana2)
print("La moda es = ", moda2)
```

```python
#Generar dos subplot de 1 renglon por dos columnas para desplegar los 2 histogramas una al lado de la otra

fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize=(10, 5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

```python
plt.hist(fabrica1, bins=10, alpha=0.2, color='b', edgecolor = "black")
plt.axvline(media1, color = 'b', linestyle = 'dashed', linewidth = 2, label= 'Media')
plt.axvline(mediana1, color = 'r', linestyle = 'dashed', linewidth = 2, label= 'Mediana')
plt.axvline(moda1, color = 'y', linestyle = 'dashed', linewidth = 2, label= 'Moda')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Producción diaria de la fabrica 1")
plt.legend()
plt.show()
```

```python
plt.hist(fabrica2, bins=10, alpha=0.2, color='b', edgecolor = "black")
plt.axvline(media2, color = 'b', linestyle = 'dashed', linewidth = 2, label= 'Media')
plt.axvline(mediana2, color = 'r', linestyle = 'dashed', linewidth = 2, label= 'Mediana')
plt.axvline(moda2, color = 'y', linestyle = 'dashed', linewidth = 2, label= 'Moda')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Producción diaria de la fabrica 2")
plt.legend()
plt.show()
```
