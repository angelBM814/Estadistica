#Parte 1
###Cargar el Dataset de Autos y realizar las Estadsticas descriptivas bsicas y medidas de tendencia central

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
from statistics import *
```

```python
autos = pd.read_csv("autos2.csv", encoding = "ISO-8859-1")

#Esta linea de cdigo elimina todos los elemetos NAN que contiene la  columna Precios, por lo que ser factible usarla para calcular el promedio
# y la desviacin estandar
precios = autos['price'].dropna()
```

```python
#Determinar el valor mnimo de precios
autos.price.min()
```

```python
#Determinar el valor mximo de precios
autos.price.max()
```

```python
#Determinar el valor de la moda de precios
modaPrecios = mode(autos.price)
print("La moda de los precios es: ", modaPrecios)
```

```python
#Determinar el valor de la media de precios
mediaPrecios = sum(precios) / len(autos.price)
print("La media de los precios es: ", mediaPrecios)
```

```python
#Determinar el valor de la mediana de precios
medianaPrecios = median(autos.price)
print("La mediana de los precios es: ", medianaPrecios)
```

```python
#Determinar el valor de la desviacin estndar de precios
desvStandPrecios = stdev(precios)
print("La desviacin estndar de los precios es: ", desvStandPrecios)
```

```python
#Determinar los estadsticos y cuartiles con la funcin describe()

print("Estadisticos y cuartiles de los precios")
autos.price.describe()
```

```python
#El punto no asume si es de los precios o de manera general, as que se
#tomarn las dos vistas

print("Estadisticos y cuartiles generales")
autos.describe()
```

#Parte 2
###Realizar un Histograma de precios de los automviles

```python
#Determinar mediante la regla de sturges la cantidad optima de bins o separadores
#Poner ttulos y encabezados apropiados
#Interprete los datos y proporcione sus conclusiones
import math
print(autos.price)
```

```python
df = pd.DataFrame({'Precios': autos.price})
df
```

```python
#Determinar mediante la regla de sturges la cantidad optima de bins o separadores
n = len(autos.price)
k = int(1 + math.log2(n))
print("Valor de n = ", n , "\nNmero de intervalos = ", k)
```

```python
#Poner ttulos y encabezados apropiados
plt.hist(df['Precios'], bins = k, alpha = 0.5)
plt.title('Histograma de Precios basado en la regla de Sturges')
plt.xlabel("Precios")
plt.ylabel("Nmero de autos")
```

##Conclusiones del histograma
Muestra una gran cantidad de automviles (aproximadamente 100) de los cuales sus precios rondan de los 5,000 a los 10,000 pesos, lo cual representa la mayor cantidad de autos. Prximo a ello se encuentran al rededor de 40 autos de los cuales sus precios rondan de los 10,000 a los 15,000 pesos y por ltimo 35 autos que rondan de 15,000 a 20,000 pesos.

Estos nos deja como conclusin en cuanto a los precios que la mayora de los autos cuestan entre 5,000 a 20,000 pesos.

#Parte 3
###Realizar un diagrama de caja y bigotes que muestre la relacin estadstica entre la columna del precio del automvil (price) y tipo de vehculo (body style)

```python
#Poner ttulos y encabezados apropiados
#Interprete los datos y proporcione sus conclusiones

sns.boxplot(x='body-style', y='price', data = autos)
plt.title('Distribucin del Precio por Tipo de Vehculo', fontsize=14)
plt.xlabel('Tipo de Vehculo', fontsize=12)
plt.ylabel('Precio del Automvil', fontsize=12)
plt.show()
```

##Conclusiones de la parte 3

Los convertibles y hardtops son los vehculos ms caros, con precios que superan los 35,000 a 40,000, mientras que los hatchbacks son los ms econmicos, con precios ms bajos y concentrados en un rango reducido. Los sedanes y wagons tienen precios ms estables y medianas bajas, aunque los sedanes pueden tener valores atpicos ms altos. En general, los deportivos como los convertibles y hardtops son ms costosos, y los hatchbacks son ms accesibles.


#Parte 4

###Realizar un diagrama de caja y bigotes que muestre la relacin estadstica entre la columna del precio del automvil (price) y tipo traccin ruedas (drive wheels)

```python
#Poner ttulos y encabezados apropiados
#Interprete los datos y proporcione sus conclusiones

sns.boxplot(x='drive-wheels', y='price', data = autos)
plt.title('Distribucin del Precio por el tipo de ruedas', fontsize=14)
plt.xlabel('Tipo de Traccin en ruedas', fontsize=12)
plt.ylabel('Precio del Automvil', fontsize=12)
plt.show()
```

##Conslusiones de la parte 4

Podemos decir que los autos con traccin trasera (RWD) son los ms caros, por el hecho de tener una mediana alta y un rango de precios amplio, incluyendo valores atpicos elevados. Los autos con traccin delantera (FWD) son los ms econmicos, con una mediana baja y precios concentrados en un rango reducido. Los autos con traccin en las cuatro ruedas (4WD) tienen precios intermedios, con una mediana ms alta que los FWD y pocos valores atpicos. En general, los RWD son ms caros debido a su enfoque deportivo o de lujo, los FWD son ms econmicos y urbanos, y los 4WD tienen un costo moderado, siendo ms utilitarios o todoterreno.
