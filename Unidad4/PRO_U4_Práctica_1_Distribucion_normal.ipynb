##Práctica 1: Distribución normal

En python se pueden utilizar las siguientes funciones para tratar con una distribución normal:
⚫ norm.rvs() para calcular números aleatorios.
⚫ norm.pdf() para calcular la densidad de probabilidad.
⚫ norm.cdf() para calcular la probabilidad acumulada.
⚫ norm.ppf() para calcular el valor de la a partir de una probabilidad, es la función inversa de cdf().

###Ejercicio 1
Una revisión de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviación estandar es de 4 kilogramos.
Suponiendo que los valores se distribuyen normalmente, Hallar lo siguiente:

Generar aleatoriamente las edades de los 500 estudiantes con una media de 70 kilogramos, una desviación estandar de 4 kilogramos.
Realizar un hisrograma de las edades.
Graficar  la función de densidad de probabilidad (PDF)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

#Generar datos para la distribución normal
media = 70 #Media
desv_std = 4 #Desviación
numDatos = 500

#generamos los datos de manera aleatoria y los ponemos en una lista
datos = np.random.normal(media, desv_std, numDatos)
datos[0:50] #Mostrar los primeros 50
```

```python
# Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b')
plt.title('Distribución normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

###Fórmula de la función de densidad de probabilidad (PDF) de la distribución normal

```python
##Crear un rango de valores para crear una densidad
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, numDatos)

#Calcular la PDF de la distribución normal
pdf = 1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp(- (x - media) ** 2 / (2 * desv_std ** 2))
```

```python
pdf.size, numDatos
```

```python
#Visulizar la PDF
plt.figure(figsize = (6, 4))
plt.plot(x, pdf, label='Normal:(media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))
plt.title('Función de Densidad de Probabilidad (PDF) - Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.legend()
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.show()
```

```python
#Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b', density=True)
plt.plot(x, pdf, 'k', linewidth=2, )
plt.title('Distribución normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

##Ejercicio 2
Una revisión de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviación estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:

1. Cuantos estudiantes pesas menos de 75kg.
2. Cuantos estudiantes pesan menos de 62kg.
3. Cuantos estudiantes pesan más de 79kg.

Primero necesitamos calcular la probabilidad de que un estudiante pese menos de 75kg utilizando la función de distribución acumulativa (CDF) de la distribución normal.

###Cuantos estudiantes pesan menos de 75kg

```python
plt.figure(figsize=(6, 4))

#parametros de la distribución normal
media = 70
desv_std = 4
totalEstudiantes = 500

#VARIABLES para cálculo de la probabilidad de que un estudiante pese menos de 75 kkg
pesoLimite = 79

#Primero necesitamos calcular la probabilidad de que un estudiante pese menos
# de 75 kg utilizando la función de distribución acumulativa (CDF) de la distribución normal
probMenos75 = norm.cdf(pesoLimite, media, desv_std)
```

```python
#calculo del número de estudiantes que pensan menos de 75kg
estudiantesMenos75 = totalEstudiantes * probMenos75
print("Probabilidad de que un estudiante pese menos de 75kg: ", probMenos75)
print("Número de estudiantes que pesan menos de 75kg: ", estudiantesMenos75)
#Graficar la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= pesoLimite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estdiante pese menos de 75kg. \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso en kg')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

###Estudiantes que pesan más de 68kg y menos de 72kg

```python

# Parámetros
media = 70
desv_std = 4
total_estudiantes = 500

# Límites de interés
peso_min = 68
peso_max = 72

# Dominio para la gráfica
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

# Calcular la probabilidad de estar entre 68 y 72 kg
prob_entre = norm.cdf(peso_max, media, desv_std) - norm.cdf(peso_min, media, desv_std)
estudiantes_entre = prob_entre * total_estudiantes

# Graficar la distribución y el área de interés
plt.figure(figsize=(6, 4))
plt.plot(x, y, 'k', linewidth=2, label='Distribución Normal')
plt.fill_between(x, y, where=(x >= peso_min) & (x <= peso_max), color='skyblue', alpha=0.5,
                 label='Entre 68 y 72 kg')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución Normal: Estudiantes entre 68 y 72 kg')
plt.legend()
plt.grid(True)
plt.show()

print("Probabilidad entre 68 y 72 kg:", prob_entre)
print("Número aproximado de estudiantes entre 68 y 72 kg:", estudiantes_entre)
```
