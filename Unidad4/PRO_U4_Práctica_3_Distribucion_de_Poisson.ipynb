###Ejercicio 1
1. Encuentre la probabilidad de que no haya accidentes en este crucero de carretera durante un periodo de 1 semana.
2. Encuentre la probabilidad de que a lo sumo haya tres accidentes en esta seccin de carretera durante un periodo de 2 semanas.

```python
from scipy.stats import poisson

# Parmetros de la distribucin de Poisson
media_accidentes = 2

# Parte 1: Probabilidad de que no haya accidentes en 1 semana
probabilidad_0_accidentes = poisson.pmf(0, media_accidentes)
print("1. Probabilidad de que no haya accidentes en 1 semana:", probabilidad_0_accidentes)

# Parte 2: Probabilidad de que a lo sumo haya tres accidentes en 2 semanas
# Convertimos la medida a una media para 2 semanas
media_2_semanas = media_accidentes * 2

# Calculamos la probabilidad acumulada hasta 3 accidentes en 2 semanas
probabilidad_3_o_menos_accidentes = poisson.pmf(3, media_2_semanas)
print("2. Probabilidad de que a lo sumo haya 3 accidentes en 2 semanas:", probabilidad_3_o_menos_accidentes)
```

###Ejercicio 2
1. Una compaa de seguros de vida asegura las vidas de 5,000 hombres de 42 aos de edad.
2. Si estudios actuariales muestran que la probabilidad de que cualquier hombre de 42 aos muera en un ao determinado es .001, encuentre la probabilidad exacta de que la compaia tendr que pagar x = 4 reclamaciones durante un ao determinado.

```python
from scipy.stats import poisson

# Parmetros de la distribucin de Poisson
media_muertes = 5000 * 0.001 # 5000 hombres con una probabilidad de muerte de 0.001

# Nmero de reclamaciones
x= 4

# Calcular la probabilidad de que la compaia tenga que pagar 4 reclamaciones en un ao determinado
probabilidad_4_reclamaciones = poisson.pmf(x, media_muertes)

print("La probabilidad de que la compaa pague", x, "reclamaciones en un ao determinado es:", probabilidad_4_reclamaciones)
```
